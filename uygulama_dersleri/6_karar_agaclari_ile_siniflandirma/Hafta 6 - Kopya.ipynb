```python
import numpy as np
import pandas as pd
from sklearn import datasets
```

```python
iris = datasets.load_iris()
```

```python
iris
```

```python
X = iris.data
y = iris.target
```

```python
X.shape
```

```python
y
```

```python
from sklearn import preprocessing
```

```python
scaler = preprocessing.StandardScaler()
```

```python
scaler.fit(X)
```

```python
scaler.mean_
```

```python
scaler.var_
```

```python
X = scaler.transform(X)
```

```python
X
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,
                                                    y,
                                                    test_size=0.3,
                                                    random_state=0)
```

```python
X_train
```

```python
X_test
```

```python
y_train.shape
```

```python
X_test.shape
```

```python
y_test.shape
```

```python
# KARAR AĞACI
from sklearn import tree
```

```python
karar_agaci = tree.DecisionTreeClassifier(criterion='entropy', 
                                          random_state=0)
```

```python
karar_agaci.fit(X_train, y_train)
```

```python
from sklearn import metrics
```

```python
dogruluk = metrics.accuracy_score(y_train, karar_agaci.predict(X_train))
```

```python
dogruluk
```
